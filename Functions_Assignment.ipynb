The terms function and method in Python both refer to callable entities, but there are key differences between them:

2. Method
A method is similar to a function but is associated with an object (usually a class instance). Methods are defined inside a class and typically operate on the instance or class data.

Methods implicitly take the instance (self) or class (cls) as their first parameter.

Example:

    def __init__(self, name):
        self.name = name

    # Method of the class
    def greet(self):
        return f"Hello, {self.name}!"

# Creating an instance of the class
greeter = Greeter("Alice")

# Calling the method
print(greeter.greet())  # Output: Hello, Alice!

Generators in Python
Generators are a special type of iterable in Python, used to generate a sequence of values lazily (on-demand). They are memory-efficient and particularly useful for handling large datasets or infinite sequences.

Key Features of Generators
Yield Instead of Return:
Generators use the yield statement to produce values one at a time, instead of returning them all at once like a regular function.
State Retention:
The generator function retains its state between successive calls, allowing it to resume where it left off.
Lazy Evaluation:
Generators do not compute all values at once; they generate values on-the-fly, saving memory.

def my_generator():
    yield 1
    yield 2
    yield 3

# Using the generator
gen = my_generator()

print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
# print(next(gen))  # Raises StopIteration

Generators in Python offer several advantages over regular functions, especially when it comes to memory efficiency, lazy evaluation, and handling large or infinite data streams. Here’s a detailed look at the advantages with examples:

1. Memory Efficiency
Advantage: Generators produce values lazily, meaning they generate one value at a time and do not store the entire dataset in memory. This makes them highly memory-efficient for working with large datasets.

2. Lazy Evaluation
Advantage: Generators evaluate and produce values only when needed, which improves performance and avoids unnecessary computation.

3. Ability to Handle Infinite Sequences
Advantage: Generators can represent infinite sequences without consuming infinite memory.

4. Improved Performance in Pipelines
Advantage: Generators are efficient when used in data pipelines, as they process one item at a time and pass it to the next step without waiting for the entire dataset to be generated.

Example:
def read_large_file(file_name):
    with open(file_name, "r") as file:
        for line in file:
            yield line.strip()

def process_lines(lines):
    for line in lines:
        yield line.upper()

lines = read_large_file("large_file.txt")
processed = process_lines(lines)

for line in processed:
    print(line)  # Processes one line at a time

Lambda Functions in Python
A lambda function in Python is an anonymous function (a function without a name) that is defined using the lambda keyword. It can take any number of arguments but can have only a single expression. The result of this expression is automatically returned.

When to Use Lambda Functions
Lambda functions are typically used in scenarios where:

A small, one-time, or throwaway function is needed.
The function is passed as an argument to higher-order functions (e.g., map, filter, sorted).
Defining a regular function would be unnecessary or make the code less readable.

Example

# Regular function
def add(x, y):
    return x + y

# Lambda function equivalent
add_lambda = lambda x, y: x + y

print(add(3, 5))        # Output: 8
print(add_lambda(3, 5))  # Output: 8

The map() Function in Python
The map() function in Python is a built-in function used to apply a given function to each item in an iterable (e.g., list, tuple, set) and return a map object (an iterator). It is a powerful tool for functional programming in Python, enabling transformations over data.

Key Features of map()
Applies a Function to Each Item: The function is called on every item in the iterable.
Supports Multiple Iterables: If the function takes multiple arguments, map() can take multiple iterables and pass corresponding items as arguments.

When to Use map()
Simple Transformations: When you need to apply a function to all elements of an iterable.
When Using Built-In or Lambda Functions: It’s ideal for simple operations where defining a full loop is unnecessary.
For Multiple Iterables: When processing corresponding elements of multiple sequences.

Example

# Function to square a number
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4]
result = map(square, numbers)

# Convert map object to a list
print(list(result))  # Output: [1, 4, 9, 16]

Differences Between map(), reduce(), and filter() Functions in Python
The functions map(), reduce(), and filter() are all higher-order functions in Python that allow you to process and transform iterables in a functional programming style. They have distinct purposes and use cases.

Summary of Usage
map(): Use when you want to transform every element in an iterable.
filter(): Use when you want to select only those elements in an iterable that meet a certain condition.

def is_prime(n):
    # Handle edge cases for numbers less than 2
    if n < 2:
        return False
    
    # Check divisibility from 2 to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # Divisible by i, so not prime
    
    return True  # No divisors found, so it's prime

# Example usage: Check prime numbers from 1 to 200
prime_numbers = [n for n in range(1, 201) if is_prime(n)]
print(prime_numbers)

class FibonacciIterator:
    def __init__(self, n):
        # Initialize the number of terms (n)
        self.n = n
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Counter to track the number of terms generated

    def __iter__(self):
        # The object itself is the iterator
        return self

    def __next__(self):
        # Stop iteration if the number of terms has been reached
        if self.count >= self.n:
            raise StopIteration
        
        # The current Fibonacci number is 'a'
        fibonacci_number = self.a
        
        # Generate the next Fibonacci number
        self.a, self.b = self.b, self.a + self.b
        
        # Increment the counter
        self.count += 1
        
        return fibonacci_number

# Example usage:
fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for num in fib:
    print(num)
